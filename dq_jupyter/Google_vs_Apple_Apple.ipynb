```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = pd.read_csv("AppleStore.csv")
data2 = pd.read_csv("googleplaystore.csv")

apple_data = pd.DataFrame()
apple_data["name"] = data1["track_name"]
apple_data["size"] = data1["size_bytes"]/1000000
apple_data["price"] = data1["price"]
apple_data["reviews"] = data1["rating_count_tot"]
apple_data["rating"] = data1["user_rating"]
apple_data["category"] = data1["prime_genre"]
apple_data["profit"] = apple_data["price"]*apple_data["reviews"]
print("Apple Total")
apple_data.to_csv("apple_clean.csv")
apple_data.describe()
```

```python
print("Apple Free")
apple_free = apple_data[apple_data["price"]==0]
apple_free.describe()
```

```python
print("Apple Paid")
apple_paid = apple_data[apple_data["price"]!=0]
apple_paid.describe()
```

```python
google_data = pd.DataFrame()
google_data["name"] = data2["App"]
google_data["size"] = (data2["Size"].str.replace("M","")).apply(pd.to_numeric,errors='coerce')
google_data["price"] = (data2["Price"].str.replace("$","")).apply(pd.to_numeric,errors='coerce')
google_data["reviews"] = data2["Reviews"].apply(pd.to_numeric,errors='coerce')
google_data["rating"] = data2["Rating"]
google_data["cateogory"] = data2["Category"]
google_data["installs"] = data2["Installs"].str.replace("+","").str.replace(",","").apply(pd.to_numeric,errors='coerce')
google_data["profit"] = google_data["price"]*google_data["reviews"]

#dropping duplicates
google_data = google_data.drop_duplicates(subset='name',keep='first')
print("Google Total")
google_data.to_csv("google_clean.csv")
google_data.describe()
```

```python
print("Google Free")
google_free = google_data[google_data["price"]==0]
google_free.describe()
```

```python
print("Google Paid")
google_paid = google_data[google_data["price"]!=0]
google_paid.describe()
```

```python
plt.pie([len(apple_free.index),len(apple_paid.index)],autopct='%1.0f%%',colors = ["green","yellow"],explode= (0.2,0),shadow = True)
plt.title("Apple free app count vs paid app count");
```

```python
plt.pie([apple_free.reviews.sum(),apple_paid.reviews.sum()] ,autopct='%1.0f%%',colors = ["green","yellow"],explode= (0.2,0),shadow = True)
plt.title("Apple free app review count vs paid app review count");
```

```python
df = apple_paid.groupby("price")["profit"].count()
plt.scatter(df.index.values,df)
```

```python
bins = [0, 0.99, 2.99, 4.99, 10.99, 14.99, 29.99]
labels = ['Under 0.99', '0.99-2.99', '2.99 - 4.99', '4.99-10.99', '10.99-14.99', '14.99-29.99']
apple_paid["price_range"] = pd.cut(apple_paid["price"],bins = bins, labels=labels)
```

```python
df = apple_paid.groupby("price_range")["profit"].count()
plt.bar(df.index.values,df)
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Apple Paid Apps - Price vs Count")
plt.xticks(rotation = 90);
```

```python
df = apple_paid.groupby("price_range")["profit"].mean()
plt.bar(df.index.values,df)
plt.xlabel("Price")
plt.ylabel("AVG Profit")
plt.title("Apple Paid Apps - Price vs Profit")
plt.xticks(rotation = 90);
```

```python
df = apple_paid.groupby("price_range")["reviews"].mean()
plt.bar(df.index.values,df)
plt.xlabel("Price")
plt.ylabel("AVG # of Reviews")
plt.title("Apple Paid Apps - Price vs Reviews")
plt.xticks(rotation = 90);
```

```python
df = apple_paid.groupby("category")["profit"].mean().sort_values(ascending=False)
plt.bar(df.index.values,df)
plt.xlabel("Category")
plt.ylabel("AVG profit")
plt.title("Apple Paid Apps - Category vs profit")
plt.xticks(rotation = 90);
```

```python
df = apple_paid.groupby("category")["reviews"].mean().sort_values(ascending=False)
plt.bar(df.index.values,df)
plt.xlabel("Category")
plt.ylabel("AVG # of Reviews")
plt.title("Apple Paid Apps - Category vs reviews")
plt.xticks(rotation = 90);
```
